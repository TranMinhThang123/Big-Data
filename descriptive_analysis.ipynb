```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Đọc dữ liệu từ file Excel
matrix = pd.read_excel("raw_data/Facebook_Data.xlsx", header=None)

# Loại bỏ dòng và cột chứa giá trị nan
matrix = matrix.dropna(axis=0, how='all')
matrix = matrix.dropna(axis=1, how='all')

# Lấy danh sách tên
names = matrix.iloc[:, 0].tolist()

# Tạo ma trận kề (0,1)
adjacency_matrix = matrix.iloc[1:, 1:].values

# Loại bỏ giá trị NaN
names = [name for name in names if pd.notna(name)]

# Loại bỏ ký tự '\xa0' sau mỗi tên
names = [name.replace('\xa0', '') for name in names]

# In ra danh sách tên
print("Danh sách tên:")
print(names)

# In ra ma trận kề (0,1)
print("Ma trận kề:")
print(adjacency_matrix)
```

```python
# Tính tổng số lượng bạn bè cho mỗi cá nhân (tính theo cột)
total_friends = np.sum(adjacency_matrix, axis=1)

# In ra số lượng bạn bè của mỗi cá nhân (có thể tùy chỉnh để xem một số lượng cá nhân nhất định)
print("Số lượng bạn bè của mỗi cá nhân:")
print(total_friends[:10])

# Tạo dictionary gắn tên với số lượng bạn bè
friend_dict = dict(zip(names, total_friends))

# In dictionary
print("Dictionary gắn tên với số lượng bạn bè:")
print(friend_dict)

# In ra vài dòng đầu của dictionary
print("Dictionary gắn tên với số lượng bạn bè (10 phần tử đầu tiên):")
for key, value in list(friend_dict.items())[:10]:
    print(f"{key}: {value}")
```

```python
# Tính các thông số mô tả
mean_friends = np.mean(total_friends)
std_dev_friends = np.std(total_friends)
median_friends = np.median(total_friends)
quartiles_friends = np.percentile(total_friends, [25, 50, 75])
min_friends = np.min(total_friends)
max_friends = np.max(total_friends)

# Hiển thị kết quả
print(f'Trung bình (mean): {mean_friends}')
print(f'Độ lệch chuẩn (standard deviation): {std_dev_friends}')
print(f'Trung vị (median): {median_friends}')
print(f'Tứ phân vị (quartiles): Q1 = {quartiles_friends[0]}, 
                                Q2 (median) = {quartiles_friends[1]}, 
                                Q3 = {quartiles_friends[2]}')
print(f'Giá trị nhỏ nhất (Min): {min_friends}')
print(f'Giá trị lớn nhất (Max): {max_friends}')
```

```python
# Chọn 10 người đầu tiên
top_10_names = list(friend_dict.keys())[:10]
top_10_friends = [friend_dict[name] for name in top_10_names]

# Vẽ biểu đồ
plt.figure(figsize=(10, 6))
plt.bar(top_10_names, top_10_friends, color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Tên người')
plt.ylabel('Số lượng bạn bè')
plt.title('Số lượng bạn bè của 10 người đầu tiên')
plt.show()
```

```python
total_friends_network = np.sum(total_friends)
print("Tổng số lượng bạn bè trên toàn bộ mạng:", total_friends_network)
```

```python
average_relationship_ratio = total_friends_network / len(names)
print("Tỷ lệ mối quan hệ bạn bè trung bình:", average_relationship_ratio)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from community import community_louvain  
from networkx.algorithms.community import greedy_modularity_communities
```

```python
def plot_subgraph(G, nodes, pos, title):
    subgraph = G.subgraph(nodes)
    plt.figure(figsize=(8, 6))
    nx.draw(subgraph, pos, with_labels=True, font_weight='bold')
    plt.title(title)
    plt.show()
```

```python
G = nx.read_edgelist('data/train.csv', delimiter=',')
```

```python
# Tính toán hệ số gom cụm trung bình
average_clustering_coefficient = nx.average_clustering(G)
print("Hệ số gom cụm trung bình:", average_clustering_coefficient)
```

```python
# Clauset Newman Moore (CNM) algorithm
communities_cnm = list(greedy_modularity_communities(G))
# Vẽ từng cụm cộng đồng trên một bảng riêng biệt
for i, community_nodes in enumerate(communities_cnm):
    plot_subgraph(G, community_nodes, nx.spring_layout(G), f'Community {i + 1}')
```

```python
# In ra các mô-đun với tên người thay vì số
print("Các mô-đun trong mạng:")
for i, comm in enumerate(communities_cnm, 1):
    valid_indices = [int(x) for x in comm if 0 <= int(x) < len(names)]
    names_in_community = map(lambda x: names[x], valid_indices)
    print(f"Mô-đun {i}: {list(names_in_community)}")
```

```python
# Tính toán độ trung gian cho từng nút
betweenness_centrality = nx.betweenness_centrality(G)

# In ra độ trung gian của mỗi nút
print("Độ trung gian của mỗi nút:")
for node, betweenness in betweenness_centrality.items():
    print(f"{node}: {betweenness}")
```

```python
# Sắp xếp các nút theo độ trung gian giảm dần và chọn 10 nút đầu tiên
top_nodes = sorted(betweenness_centrality, key=betweenness_centrality.get, reverse=True)[:10]

# In ra 10 người có độ trung gian cao nhất và gắn tên cho họ
print("10 người có độ trung gian cao nhất:")
for node in top_nodes:
    name = names[int(node)]  # Chuyển đổi chỉ số nút thành tên người
    print(f"{name}: {betweenness_centrality[node]}")
```
