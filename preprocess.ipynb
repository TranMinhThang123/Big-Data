```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("raw_data/Facebook_Data.xlsx")
df.head()
```

```python
adj_matrix = df.values[:,1:]
row,col = adj_matrix.shape
pair_matrix = []
for i in range(row):
    for j in range(col):
        if adj_matrix[i][j] == 1:
            pair_matrix.append([i+1,j+1])
csv_df = pd.DataFrame(data=np.array(pair_matrix))
csv_df.to_csv('data/train.csv',header=False,index=False)
```
